```python
from sklearn.preprocessing import StandardScaler
scaler_x = StandardScaler()
#scaler_y = StandardScaler()
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train_scaled = scaler_x.fit_transform(X_train)
#y_train_scaled = scaler_y.fit_transform(y_train)
X_test_scaled = scaler_x.transform(X_test)
#y_test_scaled = scaler_y.transform(y_test)
```

```python
y_train_mean = statistics.mean(y_train)
y_train_stdev = statistics.stdev(y_train)
y_train_scaled = (y_train-y_train_mean)/y_train_stdev 
y_test_scaled = (y_test-y_train_mean)/y_train_stdev
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train_scaled,y_train_scaled)
from sklearn.metrics import r2_score
y_pred = lin_reg.predict(X_train_scaled)
r2_score(y_pred = y_pred,y_true = y_train_scaled)
```

```python
'''Mean of residuals'''
import numpy as np
residuals = y_train_scaled-y_pred
mean_residuals = np.mean(residuals)
print("Mean of Residuals {}".format(mean_residuals))
```

```python
'''
Very clsoe to zero so all good as of now
'''
p = sns.scatterplot(x = y_pred,y = residuals)
plt.xlabel('y_pred/predicted values')
plt.ylabel('Residuals')
plt.ylim(-10,10)
plt.xlim(0,26)
p = sns.lineplot(x = [0,26],y = [0,0],color='blue')
p = plt.title('Residuals vs fitted values plot for homoscedasticity check')
```
